# Limits and Continuity

**Definition of a Limit of a function**: Let $f:D \rightarrow \mathbb{R}$ be a real valued function on $D âŠ† \mathbb{R}^d$. Then $f$ is said to have a limit $L\in \mathbb{R}$ as **$x$** approaches $a$ if: for any $\epsilon >0$, there exists a $\delta >0$ such that $|f(x)-L|<\epsilon$ for all $x \in D \cap B_\delta(a)\backslash \{a\}$. This is written as $$\lim_{x\rightarrow a}f(x)=L$$

**Definition of a continuous function**: Let $f: D\rightarrow \mathbb{R}$ be a real-valued function on $D\subseteq \mathbb{R}^d$. Then $f$ is said to be continuous at $a \in D$ if $$\lim_{x\rightarrow a} f(x)=f(a)$$

**Extreme Value Theorem**: Let $f: D\rightarrow \mathbb{R}$ be a real-valued, continuous function on a nonempty, closed, bounded set $D\subseteq \mathbb{R}^d$. Then $f$ attains a maximum and minimum on $D$.

# Derivatives

**Definition of a Derivative**: Let $f:D\rightarrow \mathbb{R}$ where $D \subseteq \mathbb{R}$ and let $x_0 \in D$ be an interior point of $D$. The derivative of $f$ at $x_0$ is $$f'(x_0)=\frac{df(x_0)}{dx}=\lim_{h\rightarrow 0} \frac{f(x_0+h)-f(x_0)}{h}$$ provided the limit exists.

**Rolle Theorem**: Let $f:[a,b]\rightarrow \mathbb{R}$ be a continuous function and assume that its derivative exists on $(a,b)$. If $f(a)=f(b)$, then there is a $a<c<b$ such that $f'(c)=0$.

**Mean Value Theorem**: Let $f:[a,b]\rightarrow \mathbb{R}$ be a continuous function and assume that its derivative exists on $(a,b)$. Then there is $a<c<b$ such that $$f(b)=f(a)+(b-a)f'(c) \ \ \ \text{or} \ \ \ \frac{f(b)-f(a)}{b-a}=f'(c)$$

```python
import sympy as sp

##derivative of function y=3x^3-5x^2-7x+4
x = sp.symbols('x')
y = 3*x**3-5*x**2-7*x+4
yy = sp.diff(y)

print('derivative of function:')
yy
```

**Definition of a Partial Derivative**: Let $f:D\rightarrow \mathbb{R}$ where $D \subseteq \mathbb{R}^d$ and let $x_0 \in D$ be an interior point of $D$. The partial derivative of $f$ at $x_0$ with respect to $x_i$ is $$\frac{\partial f(x_0)}{\partial x_i} = \lim_{h\rightarrow 0} \frac{f(x_0+he_i)-f(x_0)}{h}$$ provided the limit exists.

**Definition of the Jacobian**: Let $f=(f_1,...,f_m):D\rightarrow \mathbb{R}^m$ where $D\subseteq \mathbb{R}^d$ and let $x_0 \in D$ be interior point of $D$ where $\frac{\partial f_j(x_0)}{\partial x_i}$ exists for all $i,j$. The Jacobian of $f$ at $x_0$ is the $d \times m$ matrix $$J_f(x_0)= \left( \begin{array}{cc} \frac{\partial f_1(x_0)}{\partial x_1} & \cdots & \frac{\partial f_1(x_0)}{\partial x_d} \\ \vdots & \ddots & \vdots \\ \frac{\partial f_m(x_0)}{\partial x_1} & \cdots & \frac{\partial f_m(x_0)}{\partial x_d} \end{array} \right) $$

```python
##Get the jacobian matrix of a series of functions

x , y = sp.symbols('x y')
f = x**4+x*y**4
g = (x**2)*(y**3) + y**2

##Computing the Jacobian
J = [[f.diff(x), f.diff(y)],[g.diff(x),g.diff(y)]]
J
```

# Taylor's Theorem

**Taylors Theorem**: Let $f:D\rightarrow \mathbb{R}$ where $D \subseteq \mathbb{R}$. Suppose $f$ has a $m$ times continuous derivative on $[a,b]$. Then $$f(b)=f(a)+(b-a)f'(a)+\frac{1}{2}(b-a)^2f"(a)+...+\frac{(b-a)^{m-1}}{(m-1)!}f^{(m-1)}(a)+R_m$$ where $R_m=\frac{(b-a)^m}{(m)!}f^{(m)}(a+\theta (b-a))$ for some $0<\theta <1$.

```python
##Plot different approximations of sin(x) Taylor Series Expansion

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

##adding another element to each iteration and plotting it
plt.figure()
for n in range(4):
    y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
    label = str(2*n+1) + 'th order'
    plt.plot(x,y, label = label)

##plot the actual function
plt.plot(x, np.sin(x), 'k', label = 'sin(x)')
plt.title('Taylor Series Approximations of sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
